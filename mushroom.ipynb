```python
df1 = df.copy()
for column in replace_dict.keys():
    for key, value in replace_dict[column].items():
        df1[column] = df1[column].replace(key, value)

df1.head()
```

```python
df1.info()
```

```python
setup, X_sample, y_sample = pt.modelling.prepare_data(train_data=df1, config_path="./mushrooms.yml")
X_sample.head()
```

```python
df1["class"].value_counts(normalize=True)
```

```python
compare_df, algo_list, model_list = pt.modelling.compare_algorithms(
    setup=setup,
    include=[
        "lr",
        "nb",
    ],
    sort="recall",
    return_models=True,
)
compare_df
```

```python
compare_df_tune, model_list = (
        pt.modelling.tune_hyperparams(
               setup=setup,
               include=["nb"],
               optimize="f1",
               n_trials=10,
               return_models=True,
        )
)
compare_df_tune
```

```python
best_feature_list = pt.modelling.reduce_feature_space(
    setup=setup,
    algorithm= "nb",
    metric="f1",
    reference_metric=compare_df_tune.iloc[0]["metric"],
    acceptable_loss=0.95,
)
```

```python
compare_df_tune, model_list = (
        pt.modelling.tune_hyperparams(
               setup=setup,
               include=["nb", "lr"],
               optimize="f1",
               n_trials=20,
               feature_list=best_feature_list,
               return_models=True,
        )
)
compare_df_tune
```

```python
compare_df, algo_list, model_list = pt.modelling.compare_algorithms(
    setup=setup,
    include=[
        "lr",
        "nb",
    ],
    sort="recall",
    return_models=True,
    feature_list=best_feature_list,
)
compare_df
```
